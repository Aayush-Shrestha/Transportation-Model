```python
import numpy as np
import pandas as pd
from IPython.display import display, HTML
```

```python
file_path = 'Transport.csv'
df = pd.read_csv(file_path, header=None)
df = df.applymap(lambda x: int(x) if pd.notna(x) and not np.isinf(x) else x)
print(df.to_string(index=False, header=None))
DF = df.copy()
```

```python
wt_matrix = df.iloc[:-1, :-1].values
df2 = pd.DataFrame(wt_matrix)
print(df2.to_string(index=False, header=None))
```

```python
demand = df.iloc[-1, :].values
demand = demand[:-1]
demand = demand.astype(int)
print(len(demand))
print(demand)
```

```python
supply = df.iloc[:, -1].values
supply = supply[:-1]
supply = supply.astype(int)
print(len(supply))
print(supply)
```

```python
if np.sum(supply) == np.sum(demand):
    print("\nTotal Supply equals Total demand")
else:
    print("\nTotal Supply is not equal to total demand")
```

```python
al_matrix = np.full_like(wt_matrix, -1)
un_matrix = np.zeros_like(wt_matrix)

print(al_matrix)
```

```python
def calculate_difference(row_values):
    sorted_values = np.sort(row_values)
    for i in range(len(sorted_values) - 1):
        difference = sorted_values[i + 1] - sorted_values[i]
        if difference != 0:
            return difference
    return 0
```

```python
condition = np.sum(supply) + np.sum(demand) != 0
x = 0
while condition:
    # Calculate row_penalty
    row_pn = np.apply_along_axis(calculate_difference, axis=1, arr=wt_matrix)
    
    # Calculate col_penalty
    col_pn = np.apply_along_axis(calculate_difference, axis=0, arr=wt_matrix)

    df['row_pn'] = np.append(row_pn, [0])

    # Add col_pn as a new row with zeros for the missing values
    df.loc[len(df)] = np.append(col_pn, [0] * (len(df.columns) - len(col_pn)))

    # Find the pivot element    
    r = c = -1
    if np.max(row_pn) >= np.max(col_pn):
        r = np.argmax(row_pn)
    elif np.max(row_pn) < np.max(col_pn):
        c = np.argmax(col_pn)

    if r == -1 and c != -1:
        r = np.argmin(wt_matrix[:, c])  
    
    elif c == -1 and r != -1:
        c = np.argmin(wt_matrix[r, :])

    # Update supply, demand, cost matrix, allocation matrix and unit matrix
    al_matrix[r, c] = wt_matrix[r, c]
    

    if supply[r] > demand[c]:
        supply[r] = supply[r]-demand[c]
        un_matrix[r,c] = demand[c]
        demand[c] = 0
        wt_matrix[:, c] = 1000
        
    elif supply[r] < demand[c]:
        demand[c] = demand[c] - supply[r]
        un_matrix[r,c] = supply[r]
        supply[r] = 0
        wt_matrix[r,:] = 1000
    else:
        un_matrix[r,c] = demand[c]
        supply[r] = 0
        demand[c] = 0
        wt_matrix[:, c] = 1000
        wt_matrix[r,:] = 1000

    # Ensure the supply and demand arrays have integer data type
    supply = supply.astype(int)
    demand = demand.astype(int)
    
    # Update values in rows 0-15 and columns 0-25 from wt_matrix
    df.iloc[:16, :26] = wt_matrix[:16, :26]
    
    # Update column 26 from the supply array (update only available elements)
    df.iloc[:len(supply), 26] = supply
    
    # Update row 16 from the demand array using a for loop
    for i in range( len(demand)):
        df.loc[16, i] = demand[i]
        
    # Print the iteration number
    x = x+1
    print("Iteration : ",x)
    
    display(HTML(df.to_html()))
    print('\n')

    # Remove the last R\row
    df = df.iloc[:-1, :]

    # Remove the last column
    df = df.iloc[:, :-1]
       
    condition = np.sum(supply) + np.sum(demand) != 0
```
