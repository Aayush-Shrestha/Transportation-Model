```python
al_matrix = np.zeros_like(wt_matrix)

# Print the zero matrix
print(al_matrix)
```

```python
def calculate_difference(row_values):
    sorted_values = np.sort(row_values)
    for i in range(len(sorted_values) - 1):
        difference = sorted_values[i + 1] - sorted_values[i]
        if difference != 0:
            return difference
    return 0  # Return the last value if all differences are 0
```

```python
# Calculate row_pn
row_pn = np.apply_along_axis(calculate_difference, axis=1, arr=wt_matrix)

# Calculate col_pn
col_pn = np.apply_along_axis(calculate_difference, axis=0, arr=wt_matrix)

# Print the results
print("row_pn:", row_pn)
print("col_pn:", col_pn)
```

```python
df['row_pn'] = np.append(row_pn, [0])

# Add col_pn as a new row with zeros for the missing values
df.loc[len(df)] = np.append(col_pn, [0] * (len(df.columns) - len(col_pn)))
```

```python
print(df.to_string(index=False, header=None))
```

```python
r = c = -1
if np.max(row_pn) >= np.max(col_pn):
    r = np.argmax(row_pn)
elif np.max(row_pn) < np.max(col_pn):
    c = np.argmax(col_pn)
```
